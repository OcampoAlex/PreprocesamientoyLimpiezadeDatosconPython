```python
import pandas as pd
```

```python
# Lectura desde una petición web https
df_titanic = pd.read_csv("https://raw.githubusercontent.com/omarmendoza564/datos/main/datos/titanic.csv")
df_titanic
```

#####**LIMPIEZA DE DATOS**

1. Si la edad es null (na) llenar el campo con la media de la edad (usar el metodo fillna())

```python
mean_age = df_titanic['age'].mean()
print("La media de la edad es: ", mean_age)
df_titanic['age'] = df_titanic['age'].fillna(mean_age)
df_titanic
```

2. Si embarqued es na llenar con S

```python
df_titanic['embarked'] = df_titanic['embarked'].fillna('S')
df_titanic
```

3. Cambiar los valores alfanumericos de sex a numero usando { 'female': 1, 'male': 0 }

```python
df_titanic['sex'] = df_titanic['sex'].replace({ 'female': 1, 'male': 0 })
df_titanic
```

4. Cambiar los valores alfanumericos de embarked a numero usando { 'S': 0, 'C': 1, 'Q': 2 }

```python
df_titanic['embarked'] = df_titanic['embarked'].replace({ 'S': 0, 'C': 1, 'Q': 2 })
df_titanic
```

5. Crear una nueva variable llamada child con valor 1 para los pasajeros con una edad menor a 16 años y 0 si el pasajero tiene 16 años o mas

```python
# Ocupamos lambda para aplicar operaciones o transformaciones a los datos en una forma más compacta.
df_titanic['child'] = df_titanic['age'].apply(lambda x: 1 if x < 16 else 0)
df_titanic
```

6. Eliminar las columnas ticket, cabin y name

```python
df_titanic = df_titanic.drop(['ticket', 'cabin', 'name'], axis=1)
df_titanic
```

7. Validar que no existan registros duplicados

```python
# Buscar Dulicados considerando todos los campos
df_titanic.duplicated()
```

8. Listar los valores unicos del campo sex

```python
df_titanic['sex'].unique()
```

9. Quitar los espacios vacios de los campos que no sean numericos

```python
#Quitar caracteres en blanco
#df_titanic.info()
df_titanic['home.dest'] = df_titanic['home.dest'].str.strip()
df_titanic
```
